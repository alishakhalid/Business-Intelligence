```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("/content/drive/My Drive/Shopping Behavior Dataset (2).csv")
```

```
Q1 = df["Purchase Amount (USD)"].quantile(0.25)
Q3 = df["Purchase Amount (USD)"].quantile(0.75)
IQR = Q3 - Q1

outlier_mask = (
    (df["Purchase Amount (USD)"] < Q1 - 1.5 * IQR) |
    (df["Purchase Amount (USD)"] > Q3 + 1.5 * IQR)
)

# Remove outliers
df = df[~outlier_mask]
```

```
crosstab_df = pd.crosstab(df['Shipping Type'], df['Frequency of Purchases'])
print("Cross-tabulation of Shipping Type and Frequency of Purchases:\n")
display(crosstab_df.style.set_table_styles([
    {'selector': 'th', 'props': [('background-color', 'white'), ('color', 'black')]},
    {'selector': 'td', 'props': [('background-color', 'white'), ('color', 'black')]}
]))
```

```
crosstab_df.plot(kind='bar', figsize=(12, 7), width=0.8)
plt.title('Distribution of Frequency of Purchases by Shipping Type')
plt.xlabel('Shipping Type')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Frequency of Purchases', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Summary of the Relationship between Shipping Type and Frequency of Purchases

The cross-tabulation table and the grouped bar chart illustrate the distribution of different 'Frequency of Purchases' categories across various 'Shipping Type' options. Each bar in the chart represents the count of purchases for a specific frequency within a given shipping type.

From the **cross-tabulation table (`crosstab_df`)**, we can observe the raw counts of customers for each combination. For example, 'Free Shipping' has the highest number of 'Annually' purchases (112), while 'Next Day Air' shows a high count for 'Every 3 Months' (109). 'Standard' shipping also has a notable number of 'Annually' and 'Fortnightly' purchases.

The **grouped bar chart** visually reinforces these patterns, making it easier to compare frequencies across shipping types. We can see that:

*   **Free Shipping** appears to be quite popular across several purchase frequencies, notably for 'Annually' and 'Weekly' purchases.
*   **Express** and **Next Day Air** show relatively even distribution across frequencies, with some slight variations.
*   **2-Day Shipping** and **Store Pickup** also exhibit a fairly consistent pattern across different purchase frequencies.
*   There isn't one single 'Frequency of Purchases' that overwhelmingly dominates for all 'Shipping Type' categories. Instead, there are subtle preferences.

### Insights:

1.  **Dominance of Free Shipping**: Free Shipping tends to have higher purchase counts for 'Annually' and 'Weekly' purchases, suggesting that customers opting for these frequent or less frequent (but perhaps larger) purchase cycles might be more inclined towards cost-saving shipping options.
2.  **Varied Preferences**: Different shipping types attract customers with varied purchase frequencies. For instance, 'Next Day Air' has a relatively high number of 'Every 3 Months' purchases, indicating that customers in a hurry might not always be the most frequent buyers, but prioritize speed when they do buy.
3.  **No Strong Correlation**: While there are some discernible patterns, there isn't a universally strong correlation between a specific shipping type and a particular frequency of purchase. All shipping types seem to cater to a mix of purchase frequencies.
4.  **Strategic Opportunities**: Understanding these distributions can help retailers tailor promotions. For example, offering incentives for 'Free Shipping' to 'Annually' or 'Weekly' purchasers might further boost those segments.

### Conclusion:

The relationship between 'Shipping Type' and 'Frequency of Purchases' is nuanced. While 'Free Shipping' shows some prominence for certain purchase frequencies, customers utilize various shipping options regardless of how often they purchase. This suggests that factors other than purchase frequency, such as urgency, product type, or perceived value, likely play a significant role in shipping type selection. The data indicates a diverse customer base with different needs for both purchase frequency and shipping speed/cost.
